Question 12. A What is the purpose of the describe() function in Pandas?escribe() method in Pandas is used to generate descriptive statistics of DataFrame columns. It gives a quick summary of key statistical metrics like mean, standard deviation, percentiles, and more. By default, describe() works with numeric data but can also handle categorical data, offering tailored insights based on data type.

Syntax: DataFrame.describe(percentiles=None, include=None, exclude=None)


Parameters:


percentiles: A list of numbers between 0 and 1, specifying which percentiles to return. The default is None, which returns the 25th, 50th, and 75th percentiles.
include: A list of data types to include in the summary. You can specify data types such as int, float, object (for strings), etc. The default is None, meaning all numeric types are included.


exclude: A list of data types to exclude from the summary. This parameter is also None by default, meaning no types are excluded.


The describe() method returns a statistical summary of the data frame or series.

```python
#2 A Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

data = {"Name": ["Yash", "Kritika", "Ram", "Raj"],
        "Age": [25, 30, 37, 40],
        "Salary": [50000, 60000, 70000, 80000]}

df = pd.DataFrame(data)
mean_age = df["Age"].mean()
print("Mean of Age column:", mean_age)
```

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "A": [1, 2, 3, 4, 5],
    "B": [5, 4, 3, 2, 1],
    "C": [2, 3, 4, 5, 6],
    "D": [6, 7, 4, 3, 1]
}
df = pd.DataFrame(data)

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f')

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#5 A Generate a bar plot using Plotly
import plotly.express as px
import pandas as pd

data = {
    "Type": ["A", "B", "C", "D"],
    "Value": [10, 20, 30, 40]
}
df = pd.DataFrame(data)
fig = px.bar(df, x="Type", y="Value", title="Bar Plot")
fig.show()
```

```python
#7 Write a program to perform element-wise multiplication of two NumPy arrays
import numpy as np

arr1 = np.random.randint(1, 4, (3,3))
arr2 = np.random.randint(1, 3, (3,3))
print("array 1 :",arr1)
print("array 2 :", arr2)

multiply = arr1 * arr2
print("element wise multiplication :", multiply)
```

```python
#8 Create a line plot with multiple lines using MatplotlibA
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
y1 = x * 2   # Line 1 (y = 2x)
y2 = x * 3   # Line 2 (y = 3x)
y3 = x ** 2  # Line 3 (y = x^2)
plt.plot(x, y1, label="y = 2x", linestyle='--')
plt.plot(x, y2, label="y = 3x", linestyle='-.')
plt.plot(x, y3, label="y = x^2", linestyle=':')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```

```python
#10 Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


data = np.random.randn(500)
sns.histplot(data, bins=20, kde=True, color="red")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
#11  Perform matrix multiplication using NumPy
import numpy as np

arr1 = np.random.randint(1, 4, (3,3))
arr2 = np.random.randint(1, 3, (3,3))
print("array 1 :",arr1)
print("array 2 :", arr2)

multiply = arr1 @ arr2
print("matrix multiplication :", multiply)
```
